```python
def merge_trace(trace_id, num_traces=25):
    combined_trace = []
    
    for i in range(num_traces):
        random_trace = random.choice(traces[trace_id])
        combined_trace.extend(random_trace)
        
    return np.asarray(combined_trace)
```

```python
def unison_shuffled_copies(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]

def distance_to_shapelet(data, shapelet):
    """
    Compute the minimum distance beteen each data sample and a shapelet of interest
    """
    #data = np.asarray(data)
    #print(len(data))
    
    X = np.empty(len(data))
    for i in range(len(data)):
        D = stumpy.mass(shapelet, data[i])
        X[i] = D.min()

    return X.reshape(-1, 1)
```

```python
sample_size = 250

X = random.sample(traces[class1],sample_size)
X.extend(random.sample(traces[class2],sample_size))

y = [class1] * 250 + [class2] * 250
```

```python
X = [np.asarray(trace).astype('float64') for trace in X]
X = [trace[~np.isnan(trace)] for trace in X]

removals = [i for i,x in enumerate(X) if len(x) < 500]

for idx in removals:
    X[idx] = None
    y[idx] = None

X = [trace for trace in X if trace is not None]
y = [value for value in y if value is not None]

#X, y = unison_shuffled_copies(np.asarray(X, dtype=object), np.asarray(y, dtype=object))
X_train_org, X_test_org, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
clf = DecisionTreeClassifier()

for i, (gun_shapelet, point_shapelet) in enumerate(zip(gun_shapelets, point_shapelets)):
    X_train_gun = distance_to_shapelet(X_train_org, gun_shapelet)
    X_train_point = distance_to_shapelet(X_train_org, point_shapelet)
    X_train = np.concatenate((X_train_gun, X_train_point), axis=1)
    
    X_test_gun = distance_to_shapelet(X_test_org, gun_shapelet)
    X_test_point = distance_to_shapelet(X_test_org, point_shapelet)
    X_test = np.concatenate((X_test_gun, X_test_point), axis=1)
    
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f"Accuracy for shapelet {i} = {round(metrics.accuracy_score(y_test, y_pred), 5)}")
```
